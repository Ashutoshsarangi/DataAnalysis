## Q2. Remove all the Records where confirmed cases less than 10 .

```python
df.drop(df[df['Confirmed'] < 10].index, inplace=True)
```

```python
df.shape
```

## Q3. Which region maximun Number of Cases recorded?

```python
df[['Region','Confirmed']][df['Confirmed'] == df['Confirmed'].max()]
```

## Q4. Which region minimun Deaths were recorded 

```python
df.columns
```

```python
df[['Region','Deaths']][df.Deaths == df.Deaths.min()]
```
