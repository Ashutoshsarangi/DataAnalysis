temp = {
'temprature':[23, 15, -9999, 34, -999],
'windspeed':[6, -8888, 8,-8888, 9],
'event':['sunny', 0, 'rain', 'cold', 0]
}



```python
import pandas as pd

df = pd.DataFrame( { 'temprature':['23', '15', '-9999', '34', '-999'], 'windspeed':['6', '-8888', '8','-8888', '9'], 'event':['sunny', 0, 'rain', 'cold', 0] })
```

```python
df
```

```python
# Now time to Replace all Abnormal Entries
import numpy as np
new_df = df.replace({
    'temprature':['-9999', '-999'],
    'windspeed':['-8888', '-8888'],
    'event':0
}, np.NaN)

new_df
```
