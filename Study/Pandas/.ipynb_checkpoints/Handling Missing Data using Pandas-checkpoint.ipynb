There are actually 3 methods are there for the Handlig Missing Datas

- fillna()
    - {} #Dictonary
    - ffill
    - bfill
- interpolate()
    - time
- dropna()
    - how
    - thresh

```python
import pandas as pd

df = pd.read_excel('weather_data.xls')
```

```python
df
```

```python
# Here in FillNa we can pass the column name and give respective values when Ever there is a NAN values
df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'Not Available'
})
```

```python
#Now Its Time for Forward Fill
df.fillna(method='ffill')
```

```python
#time For Backward Fill
df.fillna(method='bfill')
```

```python
# In the Interpolate mechanism it will kind of put the interpolated valus based on the available values

df.interpolate()
```

```python
df.reset_index(inplace=True)
df.set_index('day', inplace=True)
```

```python
df.interpolate(method="time")
```

```python
#time for dropna()

df.dropna(thresh=1) #how='all' when all the values are null we need to remove those
#thresh=1 means greater than 1 column values are NAN then, remove it.
```
