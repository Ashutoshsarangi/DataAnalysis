```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5,6,7]
y= [50, 40, 35,67,60, 32, 54]
```

```python
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.title('Weather Report')
plt.plot(x,y, color="green", linewidth='2')
```
