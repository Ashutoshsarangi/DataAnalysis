```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
min_t = [30,40,34,23,28,35,45]
max_t = [50,45,40,25,40,50,60]
avg_t = [45,43,37,24,35,45,50]
```

```python
# time for plot
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x, min_t)
plt.plot(x, max_t)
plt.plot(x, avg_t)
```

```python
# Now tome to add Legend and Lebel

plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x, min_t, label ='Min')
plt.plot(x, max_t, label='Max')
plt.plot(x, avg_t, label='Avg')
plt.legend(fontsize='large', shadow=True)
```
