```python
import matplotlib.pyplot as plt
```

```python
values=[100, 200, 50, 500, 300]
labels= ['Apple', 'Milk', 'Veg', 'Chicken', 'Fish']
```

### [Click Here to get More Ideas on Pie Chart Api](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)

```python
plt.pie(values, labels=labels)
```

```python
# remove the Exta Information and Show The Proper Shape of Pie
plt.axis('equal')
plt.pie(values, labels=labels)
plt.show()
```

```python
#Now Show Case the %
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%0.2f%%') # 0.2 stands for If you want upto 2 decmal point
plt.show()
```

```python
#Now Show Case the %
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%f%%') # By Default it will Show 5/6 decimal points
plt.show()
```

```python
#Now Show Case the %
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%0.0f%%') 
plt.show()
```

```python
#You want to Explode a slice
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%0.0f%%', explode=[0,0,0,0,0]) 
plt.show()
```

```python
#You want to Explode a slice
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%0.0f%%', explode=[0,0.1,0,0,0.1]) 
plt.show()
```

```python
#lets say you want to incese the Size of the pie chart
#You want to Explode a slice
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%0.0f%%', explode=[0,0,0,0,0], radius=1.5) #radious is responsible for Size Increase.
plt.show()
```

### Now Its Time to Save the Chart in a png File

### [Click here to get More Parameters](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html)

```python
plt.axis('equal')
plt.pie(values, labels=labels, autopct='%0.0f%%', explode=[0,0,0,0,0], radius=1.5) #radious is responsible for Size Increase.

plt.savefig('pieChart.png', bbox_inches='tight')
```

## Note 

If you are using plt.show() then Saving your image it will show the Blank Image So you need to remove the Show() method to save the image in File
