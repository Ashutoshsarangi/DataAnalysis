```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['Google', 'Amazon', 'Microsoft', 'FaceBook']
revenue=[50, 75, 45, 40]
profit = [20, 10, 23, 25]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
# Vartical bar Chart
plt.ylabel('Revenue')
plt.title('Company Vs Revenue')
plt.bar(company, revenue, label='Revenue', width=0.4)
plt.bar(company, profit, label ='Profit', width=0.4)
plt.legend()
```

```python
# Now Lets Say I want a Separated Bar chart for revnue and Profit in that Scenario I must need to use xpos
```

```python
# Vartical bar Chart
plt.ylabel('Revenue')
plt.title('Company Vs Revenue')
plt.bar(xpos-0.2, revenue, label='Revenue', width=0.4)
plt.bar(xpos+0.2, profit, label ='Profit', width=0.4)
plt.legend()
```

```python
# Now in X Asis If you Notice My company Name is Gone to show case company name i have to use xticks function
plt.xticks(xpos, company)
plt.ylabel('Revenue')
plt.title('Company Vs Revenue')
plt.bar(xpos-0.2, revenue, label='Revenue', width=0.4)
plt.bar(xpos+0.2, profit, label ='Profit', width=0.4)
plt.legend()
```

```python
# Now its Time for Horizontal Bar Chart

plt.yticks(xpos, company)
plt.xlabel('Revenue')
plt.title('Company Vs Revenue vs Profit')
plt.barh(xpos-0.2, revenue, label='Revenue', height=0.4)
plt.barh(xpos+0.2, profit, label ='Profit', height=0.4)
plt.legend()
```
