```python
import matplotlib.pyplot as plt
```

```python
blood_sugar_men = [100, 80, 85, 123, 90, 87, 140, 110, 95, 90, 76, 89]
blood_sugar_women = [90, 89,78, 100, 120, 98, 85, 120, 95, 87, 90,138]
```

```python
plt.hist(blood_sugar_men) # By Default It will show 10 Bins

# but for us we need 3 bins [80-100, 100-125, 125-150] 3 range
```

```python
plt.hist(blood_sugar_men, bins=3)
```

```python
plt.hist(blood_sugar_men, bins=[80,100,125,150])
```

#### [Click here for More Detail on Hist API](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
plt.hist(blood_sugar_men, bins=[80,100,125,150], rwidth=0.95) # relative width is responsible for the Gap
```

```python
plt.hist(blood_sugar_men, bins=[80,100,125,150], rwidth=0.95, color=['green'])
```

```python
# Now time to add Both Men and Women in the Histagram Plot
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'])
```

```python
# Now adding Level into the chartas and Legends and Axis Name and Title
plt.xlabel('Sugar Lebel')
plt.ylabel('Number of Patient')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'], label=['Men', 'Women'])
plt.legend()
```

```python
# Now adding Level into the chartas and Legends and Axis Name and Title
plt.xlabel('Sugar Lebel')
plt.ylabel('Number of Patient')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], 
         rwidth=0.95, color=['green', 'orange'], label=['Men', 'Women'],
        orientation='horizontal')
plt.legend()
```
